```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Customertravel.csv")
df.head()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
data_types = df.dtypes
print(data_types)
```

```python
sns.set(style="whitegrid")
```

```python
# Distribution of the target variable
plt.figure(figsize=(8, 6))
sns.countplot(x='Target', data=df, palette='viridis')
plt.title('Distribution of Churn')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()
```

```python
# Age distribution by Churn
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Target', multiple='stack', palette='viridis')
plt.title('Age Distribution by Churn')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Annual Income Class distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='AnnualIncomeClass', hue='Target', data=df, palette='viridis')
plt.title('Annual Income Class Distribution by Churn')
plt.xlabel('Annual Income Class')
plt.ylabel('Count')
plt.show()
```

```python
# Services Opted Frequency by Churn
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='ServicesOpted', hue='Target', multiple='stack', palette='viridis')
plt.title('Services Opted Frequency by Churn')
plt.xlabel('Services Opted Frequency')
plt.ylabel('Count')
plt.show()
```

```python
# Social Media Account Synchronization by Churn
plt.figure(figsize=(8, 6))
sns.countplot(x='AccountSyncedToSocialMedia', hue='Target', data=df, palette='viridis')
plt.title('Social Media Account Synchronization by Churn')
plt.xlabel('Account Synced to Social Media')
plt.ylabel('Count')
plt.show()
```

```python
# Hotel Bookings by Churn
plt.figure(figsize=(10, 6))
sns.countplot(x='BookedHotelOrNot', hue='Target', data=df, palette='viridis')
plt.title('Hotel Bookings by Churn')
plt.xlabel('Booked Hotel or Not')
plt.ylabel('Count')
plt.show()
```

```python
churn_distribution = df['Target'].value_counts()
print(churn_distribution)
```

```python
# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(churn_distribution, labels=['Non-Churn', 'Churn'], autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#ff6666'])
plt.title('Churn Distribution')
plt.show()
```

```python
# Box plot for Age by Churn
plt.figure(figsize=(10, 6))
sns.boxplot(x='Target', y='Age', data=df, palette='viridis')
plt.title('Age Distribution by Churn')
plt.xlabel('Churn')
plt.ylabel('Age')
plt.show()
```

```python
# Count plot for FrequentFlyer by Churn
plt.figure(figsize=(10, 6))
sns.countplot(x='FrequentFlyer', hue='Target', data=df, palette='viridis')
plt.title('Frequent Flyer Status by Churn')
plt.xlabel('Frequent Flyer')
plt.ylabel('Count')
plt.show()
```

```python
# Pair plot to see pairwise relationships
sns.pairplot(df, hue='Target', palette='viridis')
plt.title('Pairwise Relationships')
plt.show()
```
